```python
import pandas as pd
try:
    df = pd.read_csv("C:/Users/91831/Downloads/THIRD YEAR!!!/Heart_Disease_Prediction.csv/Heart_Disease_Prediction.csv")
except FileNotFoundError:
    print("Error: 'MultipleFiles/Heart_Disease_Prediction.csv' not found. Please ensure the file is in the correct directory.")
    exit()
print("Dataset loaded successfully. First 5 rows:")
print(df.head())
print("\nDataset Info:")
df.info()
print("\nDataset Description:")
print(df.describe())
```

```python
import numpy as np
print("\n1) Mean, Median, Mode for numerical columns:")
for column in df.select_dtypes(include=np.number).columns:
    mean_val = df[column].mean()
    median_val = df[column].median()
    mode_val = df[column].mode()[0] # .mode() can return multiple values if there's a tie
    print(f"  Column '{column}':")
    print(f"    Mean: {mean_val:.2f}")
    print(f"    Median: {median_val:.2f}")
    print(f"    Mode: {mode_val}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
print("\n2) Outlier Detection using Box Plot/IQR Method:")
outliers_dict = {}
for column in df.select_dtypes(include=np.number).columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    column_outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    if not column_outliers.empty:
        outliers_dict[column] = column_outliers
        print(f"  Outliers found in '{column}':")
        print(column_outliers[[column]])
    else:
        print(f"  No significant outliers found in '{column}'.")
```

```python
import seaborn as sns
plt.figure(figsize=(15, 10))
for i, column in enumerate(df.select_dtypes(include=np.number).columns):
    plt.subplot(3, 5, i + 1) # Adjust subplot grid based on number of numerical columns
    sns.boxplot(y=df[column])
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```

```python
print("\n3) Kurtosis for numerical columns:")
for column in df.select_dtypes(include=np.number).columns:
    kurt = df[column].kurtosis()
    print(f"  Column '{column}': Kurtosis = {kurt:.2f}")
    if kurt > 0:
        print("    (Leptokurtic: more peaked than a normal distribution)")
    elif kurt < 0:
        print("    (Platykurtic: less peaked than a normal distribution)")
    else:
        print("    (Mesokurtic: similar peakedness to a normal distribution)")
```

```python
print("\n4) Skewness for numerical columns:")
for column in df.select_dtypes(include=np.number).columns:
    skew = df[column].skew()
    print(f"  Column '{column}': Skewness = {skew:.2f}")
    if skew > 0:
        print("    (Positively skewed: tail on the right)")
    elif skew < 0:
        print("    (Negatively skewed: tail on the left)")
    else:
        print("    (Symmetric)")
```

```python
print("\n1) Correlation Matrix for numerical variables:")
correlation_matrix = df.select_dtypes(include=np.number).corr()
print(correlation_matrix)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Convert 'Presence' to 1, 'Absence' to 0
df['Heart Disease'] = df['Heart Disease'].map({'Presence': 1, 'Absence': 0})
print("\n  Example: Correlation between 'Age' and 'Heart Disease':")
age_heartdisease_corr = df['Age'].corr(df['Heart Disease'])
print(f"  Pearson correlation coefficient: {age_heartdisease_corr:.2f}")

if age_heartdisease_corr > 0.5:
    print("  --> Strong positive correlation.")
elif age_heartdisease_corr < -0.5:
    print("  --> Strong negative correlation.")
else:
    print("  --> Weak or no correlation.")
```
